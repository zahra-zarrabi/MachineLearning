```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
```

```python
def generate_data(n):
    hair_length_female = np.random.normal(40,7,n)
    hair_length_man=np.random.normal(10,3,n)
    
    label_female=np.zeros(n,dtype='int')
    label_man=np.ones(n,dtype='int')
    
    hair_length=np.concatenate((hair_length_female,hair_length_man))
    gender=np.concatenate((label_female,label_man))
    return hair_length,gender
```

```python
n=100
X,Y=generate_data(n)
X=X.reshape(-1,1)
Y=Y.reshape(-1,1)
```

```python
plt.scatter(X[:n],Y[:n],label='famale')
plt.scatter(X[n:],Y[n:],label='man')
plt.xlabel("Hair Length")
plt.ylabel('Gender')
plt.legend()
```

```python
#plt random slope
plt.scatter(X[:n],Y[:n],label='famale')
plt.scatter(X[n:],Y[n:],label='man')
plt.xlabel("Hair Length")
plt.ylabel('Gender')
plt.legend()

for m in range(3):
    y_pred =np.matmul(X,[m])
    plt.plot(X,y_pred,c='red')
```

```python
# Adaline 

m=np.matmul(inv(np.matmul(X.T,X)),np.matmul(X.T,Y))
y_pred=np.matmul(X,m)

plt.scatter(X[:n],Y[:n],label='famale')
plt.scatter(X[n:],Y[n:],label='man')
plt.plot(X,y_pred,c='red',lw=3)
plt.xlabel('Hair Length')
plt.ylabel('Gender')
plt.legend()
```
