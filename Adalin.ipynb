```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
```

```python
def generate_data(n):
    height=np.random.normal(120,50,n)
    weight=np.random.normal(50,30,n)
    return height,weight
```

```python
height,weight=generate_data(n=100)
print(height,weight)
```

```python
X=height.reshape(-1,1)
Y=weight.reshape(-1,1)
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X,Y)
for m in range(4):
    m=np.array([height[m]/weight[m]])
    Y_pred=np.matmul(X,m)
    plt.plot(X,Y_pred,c='red')
```

```python
m=np.matmul(inv(np.matmul(X.T,X)),np.matmul(X.T,Y))
y_pred=np.matmul(X,m)
plt.scatter(X,Y)
plt.plot(X,y_pred,c='red')
```
