```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('weatherHistory.csv')
data.head(50)
```

```python
date = pd.to_datetime(data['Formatted Date'])
```

```python
date_list = [int(tmp.strftime('%j')) for tmp in date.tolist()]
degree_list = data['Temperature (C)']
import numpy as np
number_of_day_year = np.max(date_list)
day_degree = [list() for _ in range(number_of_day_year)]

for indx in range(len(date_list)):
    day = date_list[indx]
    day_degree[day-1].append(degree_list[indx])
day_degree_np = np.zeros((number_of_day_year, 1))
for indx, degrees in enumerate(day_degree):
    day_degree_np[indx] = np.mean(degrees)
```
